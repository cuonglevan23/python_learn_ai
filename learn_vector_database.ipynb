# Distance Metrics and Loss Functions
- trong vector database, như FAISS, Pinecone hay Weavite mỗi đối tượng (ảnh,văn,sản phẩm, người dùng) được ánh xạ thành một vector trong không gian nhiều chiều.
- khi cần tìm kiếm đối tương tự hệ thống ó sánh độ giống nhau hoặc độ khác biệt giữa các vector, các hàm dùng để do khoảng cách hoăch đọ tương đồng này gọi kà Distance Loss Function.

# L2 Distance (Euclidean Distance)
- công thức : L2(A,B) = sqrt(sum((A[i] - B[i])^2))
  - trong đó: A = [a1, a2, ... , an]
             B = [b1, b2, ... , bn]
    n là số chiều (số phân tử của vector).
- ví dụ: A = [1,2], B = [4,6] , L2. = sqrt((1-4)^2 + (2-6)^2) = sqrt(9 + 16) = sqrt(25) = 5 
- ứng dụng: trong tìm kiếm hình ảnh, nhận diện khuôn mặt, phù hợp khi embeddung chưa được chuẩn hóa
             

# L1 Distance (Manhattan Distance)
- công thưc : L1(A,B) = sum(|A[i] - B[i]|) trong |Ai - Bi| là khoảng cách tuyệt đối giữa hai thành phần tương ứng, n là số chiều của vector.
- Ví dụ  A = [1,2], b = [4,6] VÀ L1 = |1-4| + |2-6| = 3 + 4 = 7
- ứng dụng: dùng trong môi trường có nhiều nhiễu hoặc giá trị ngoại lai và phù hợp với dữ liệu rời rạc (discrete) hoặc không liên tục


# Cosine Similarity / Cosine Distance
- Công thức cosine similarity: cos(A,B) = (A.B) / (||A|| ||B||)
- trong đó A.B là tích vô hướng, ||A|| là ĐỘ DÀU CHUẨN của vector A và ||B|| là ĐỘ DÀU CHUẨN của vector B
- Distancecos = 1 - cos(A,B)
- ví dụ : A = [1,0] B = [0,1] thì cos(A,B) = 0 (góc 90 độ), A[1,1], B=[2,2] => cos = 1( cùng hướng)
- ứng dụng: thường dùng cho NLP, so khớp câu hỏi, tìm kíem văn bản, không bị ảnh hưởng bởi độ dài vector (chỉ quan tâm đến góc giữa hai vector) 


# Dot Product (Inner Product)
- công thức: A*B = A[0]*B[0] + A[1]*B[1] + ... + A[n]*B[n]
- trong đó: a[i] là phần tử thứ i của vector A, b[i] là phần tử thứ i của vector B, kết quả là một số vô hướng, phản ảnh mực độ "Đồng hướng và độ lớn"
- ví dụ: A = [1,2], B = [3,4] thì A*B = 1*3 + 2*4 = 11
- ứng dụng: dùng trong các hệ thống tìm kiếm với vector lớn và hiệu suất cao
- phổ biến CLIP, Bert hoặc khi dùng Approximate Nearest Neighbor (ANN)
