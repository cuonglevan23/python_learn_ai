# Cấu trúc dữ liệu chính
- Pandas xây dựng trên hai cấu trúc dữ liệu cốt lõi: Seriec và DataFrame, đây là nên tảng của mọi thao tác phân tích trong pandas

- Trong pandas, Series là một cấu trúc dữ liệu một chiều hoạt động như một mảng(array) có gán nhãn, mỗi phần tử trong series bao gồm một giá trị và một chỉ mục(index), mặc địch, nếu không chỉ định index, pandas sẽ tự động tạo ra index từ 0 đến n-1.
  + ví dụ: s = pd.Series([1, 3, 5, np.nan, 6, 8])
  + trong trường hợp này các phần từ sẽ có index là 0, 1, 2, 3, 4, 5 tương ứng với các giá trị, đây là là cách đơn giản nhất để tạo ra một series, tương tự như một mảng có đánh số dòng
  + tuy nhiên, pandas cho phép tùy chỉnh index bằng cách ruyền thêm tham số index = [....] khi khởi tạo series. Điểu này hữu ích khi muôn gán nhãn rõ ràng cho dữ liệu thay vì chỉ số số nguyên mặc định
  + ví dụ tạo series với index được gán thủ công:
  s = pd.Series([1, 3, 5, np.nan, 6, 8], index = ['a', 'b', 'c', 'd', 'e', 'f'])
  + việc gán index tùy chỉnh cho phspe truy xuất dữ liệu dễ dàng hơn bằng cách dùng nhãn thay vì số, ví dụ s['a'] sẽ trả về giá trị đầu tiên trong series  = 1.

- DataFrame là cấu trúc bảng hai chiều, gồm nhiều series, xép thành cột với nhãn dòng index và nhãn cột (columns)
  + ví dụ data = {'Name': ['John', 'Anna', 'Peter', 'Linda'], 'Age': [28, 22, 34, 42], 'City': ['New York', 'Paris', 'London', 'Berlin']}
  df = pd.DataFrame(data, columns=['Name', 'Age', 'City'])
  + mỗi cột trong dataframe có thể mang kiểu dữ liệu khác nhau, một số kiểu dữ liệu phổ biến gồm:
    + int64
    + float64
    + object
    + bool
    + datetime64
    + category

# Nhóm dữ liệu
- trong phân tích dữ liệu, việc nhóm các bản ghi theo một hoặc nhiều tiêu chí là thao tác phổ biến để tính toán thống kê cho từng nhóm riêng biệt, Pandas cung cấp phương thức groupby() để thực hiện điều này một cách hiệu quả và linh hoạt
- groupby hoạt động theo quy trình: Split dữ liệu thành từng nhóm. Apply hàm xử lý(tính trung bình, tổng, đếm ,vv.v.) và combine kết quả lại



```python
# ví dụ về groupby
data = {
 'Name': ['Anh', 'Thai', 'Hoa'],
 'Country': ['Brazil', 'Vietnam', 'Vietnam']
 }
df = pd.DataFrame(data, columns=['Name', 'Country'])
df
df.groupby('Country').agg({'Name': 'count'})
```

```python
# data salary
data = {
    "name": ["John", "Jane", "Jack", "Jill"],
    "department": ["HR", "IT", "HR", "IT"],
    "salary": [50000, 60000, 55000, 70000],
    "gender": ["Male", "Female", "Male", "Female"]
}
df = pd.DataFrame(data)
print(df)
df.groupby("department")["salary"].mean()

# nhóm nhiều cột cùng lúc
df.groupby(["department", "gender"])["salary"].sum()
```

-  áp dụng hàm trùy chỉnh
 + apply() là phương thức linh hoạt cho phép áp dung một hàm bất kỳ lên từng hàng hoặc từng cột của DataFrame hoặc series. Rất hưu ích khi cần xử lý theo quy tắc phức tạp không có sẵn trong pandas.
  -  ví dụ: chuẩn hóa dữ liệu trong cột score về thàng điểm 10: 
     - df['score'] = df['score'].apply(lambda x: x/10)
     áp dụng một hàm xử lý cho từng dòng:
     + df.apply(lambda row: row['math'] + row['english'], axis=1) tham số axis=1 cho phép áp dụng hàm cho từng dòng nếu axis=0 thì áp dụng cho từng cột

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Country': ['USA', 'Canada', 'Austria', 'Germany'],
    'Grade': [88, 92, 87, 93]
}
df = pd.DataFrame(data)
print(df)

f = lambda x: x * 2
df['Grade'] = df['Grade'].apply(f)
print(df)
```

# Rolling 
- phương thức rolling() trong pandas tạo ra một đối tượng 'cửa số trượt'(rolling window) cho phép áp dụng các hàm thống kế (như mean, sum, std,v.vv) lên các tập giá trụ con liên tiếp theo chiều dọc của một cột, đây là công cụ đặc biệt hữu ích trong các bài toán phân tích chuỗi thời gian, nơi cần quan sát các xu hướng, biến động hoặc các đặc điểm thống kê trong các khoảng thời gian nhất định.
  + cho phép áp dụng các phép biến đổi tổng hợp trên 1 phạm vi giới hạn trượt theo từng dòng 
  + Hữu ích để làm mượt dữ liệu, phát hiện xu hướng ngắn hạn, hoặc tính toán trung bình động
  + có thể tùy chỉnh kích thước cửa số, lựa chọn căn chỉnh tâm cửa số , xử lý cạnh và các thống ố khác để phù hợp với yêu cầu phân tích cụ thể 

  

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Country': ['USA', 'Canada', 'Austria', 'Germany'],
    'Grade': [88, 92, 87, 93]
}
df = pd.DataFrame(data)
print(df)

df['Grade'].rolling(3).sum() # tính tổng của mỗi cửa số 3 giá trị liên tiếp trong cột Grade. kết quả sẽ trả về một series mới, trong đó phân tử thứ ba chứ tổng của 3 giá trị đầu tiên, nếu chưa đủ dữ liệu sẽ nhận giá trị NaN
```

# Làm sạch dữ liệu
 - dữ liệu thu nhập từ thực tế hiếm khi hoàn chỉnh, chúng thường có những vấn đề phổ biến như: ô trống(missing values), định dạng sai(wrong format), nhiễu(noise), nhiều giá trị trùng lặp(duplicates), hãy dữ liệu bất hợp lý (invalid data). việc xử lý các vấn đề này là bước thiết yếu trước khi phân tích hay xấy dựng mô hình học máy, học sâu
 

```python
import pandas as pd
import numpy as np

data = {
    'name': ['John','Anna','James','Linda','Mark','John','Paul'],
    'age': [28,22,35,32,25,np.nan,22],
    'city': ['New York','Paris','London','Berlin','Tokyo','New York','Tokyo']
}

df = pd.DataFrame(data)
print(df)
```

```python
# xử lý dữ liệu NAN
df.dropna()
df.dropna(axis=1)
df.fillna(0)
df['age'].fillna(df['age'].mean())
print(df)
```

# Sửa định dạng dữ liệu 
- định dạng dữ liệu không đúng có thể gây lỗi khi tính toán hoặc trục quan hóa. Pandas hỗ trọ ép kiểu (type conversion) để chuyển đổi dữ liệu thành định dạng mong muốn. để đảm báo tính chính xác của dữ liệu

```python
data = {
    'name': ['John','Anna','James'],
    'age': ['28', '22', ''],
    'join_date': ['2022-01-01', '2021-12-15', '2023-03-10'],
    'is_active': ['True', 'False', 'True']
}

df = pd.DataFrame(data)

df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['join_date'] = pd.to_datetime(df['join_date'])
df['is_active'] = df['is_active'].astype(bool)

print(df)
```

# Xóa dữ liệu trùng lặp
- các bản ghi trùng lặp thường làm sai lệch thống kê hoặc gây thiên lệch trong mô hình. pandas hỗ trợ nhiều phương pháp hiện và loại bỏ trùng lặp linh hoạt.
- ví dụ: df.drop_duplicates(), df.drop_duplicates(subset=["column1", "column2"]) , df.drop_duplicates(keep="first") 
- giải thích từng câu lệch drop_duplicates xóa các dòng trùng lặp hoàn toàn mọi giá trị giống nhau. dùng trong các bảng dữ liệu tổng hợp hoặc xuất từ hệ thống lỗi
- subset = "email" chỉ xóa trùng lặp trong cột email nếu hai bản ghi có cùng địa chỉ email thì coi là trùng lặp
- keep = "last" chỉ xóa các bản ghi trùng lặp và giữ lại bản ghi cuối cùng
- keep = "first" chỉ xóa các bản ghi trùng lặp và giữ lại bản ghi đầu tiên
- keep = False xóa tất cả các bản ghi trùng lặp
- dùng df[df.duplicated()] để kiểm tra xem có bản ghi trùng lặp không
